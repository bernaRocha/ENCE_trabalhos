<b>Questão 2. </b>

O número de filhos em um grupo de 12 casais é:

1, 0, 2, 0, 1, 3, 2, 2, 0, 1, 2, 4.

a) Calcule a média, a moda e a mediana do número de filhos. <br>
b) Calcule a amplitude, variância e desvio padrão do número de filhos.

```python
filhos = [1, 0, 2, 0, 1, 3, 2, 2, 0, 1, 2, 4]

media =  statistics.mean(filhos)
moda = statistics.mode(filhos)
mediana = statistics.median(filhos)
variancia = statistics.variance(filhos)
desvio_padrao = statistics.stdev(filhos)

print(f'''A média de filhos por casal é: {media}, 
A moda de filhos é: {moda}
A mediana de filhos é: {mediana}
A amplitude de filhos é: {max(filhos) - min(filhos)}
A variância de filhos é: {round(variancia, 2)}
Desvio padrão de filhos: {round(desvio_padrao, 2)}
''')
```

```python
dados = [56, 61, 57, 77, 62, 75, 63, 55, 64, 60, 60, 57, 61, 57, 67, 62, 69, 67, 68, 59,
65, 72, 65, 61, 68, 73, 65, 62, 75, 80, 66, 61, 69, 76, 72, 57, 75, 68, 83, 64,
69, 64, 66, 74, 65, 76, 65, 58, 65, 64, 65, 60, 65, 80, 66, 80, 68, 55, 66, 71]

df = pd.DataFrame({'glicose_crianca': dados})
```

```python
# Na moda foi essa a solução para que não me retorne uma tupla

print(f'''Média:         {round(df.mean()[0], 3)}
Moda:          {df['glicose_crianca'].value_counts().idxmax()} 
1° decil:      {df.quantile(0.10)[0]}
4° decil:      {df.quantile(0.40)[0]}
1° percentil:  {df.quantile(0.01)[0]}
80° percentil: {df.quantile(0.80)[0]}
Amplitude:     {df.max()[0] - df.min()[0]}
Variância:     {round(df.var()[0], 3)}
Desvio padrão: {round(df.std()[0], 3)}
''')
```
